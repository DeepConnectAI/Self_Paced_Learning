### Building the classifier

#### prior probability

```python
def prior_prob(y_train,label):
    numerator = np.sum(y_train==label)
    denominator = y_train.shape[0]
    return (numerator)/float(denominator)
```

#### Likelihood / prosterior prob

```python
def likelihood(X_train,y_train,column,column_value,label):
    X_filtered = X_train[y_train==label]
    numerator = np.sum(X_filtered[:,column]==column_value)
    denominator = np.sum(y_train==label)
    return (numerator)/float(denominator)
```

#### predict

```python
def predict(X_train,y_train,X_test):
    prosterior_prob = []
    lables = np.unique(y_train)
    for label in lables:
        likeli = 1.0
        prior = prior_prob(y_train,label)
        for feature in range(len(X_test)): 
            cond = likelihood(X_train,y_train,feature,X_test[feature],label)
            if cond==0: 
                continue
            likeli *= cond
        prost_prob = prior*likeli
        prosterior_prob.append(prost_prob)
    return np.argmax(prosterior_prob)
```

```python
print(predict(X_train,y_train,X_test[4]))
print(y_test[4])
```

```python
def score(X_train,y_train,X_test,y_test):
    y_pred = []
    for x in range(X_test.shape[0]):
        pred = predict(X_train,y_train,X_test[x])
        y_pred.append(pred)
    return (np.sum(y_pred==y_test))/y_test.shape[0]
        
    
```

```python
score(X_train,y_train,X_test,y_test)
```
